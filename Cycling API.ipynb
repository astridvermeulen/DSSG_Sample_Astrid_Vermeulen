```python
races = pd.DataFrame()
Race_ids = [17, 13, 23, 1, 2, 3, 12, 15, 16, 14, 6, 19, 21, 9800, 4, 7, 5, 8, 9, 11, 59, 58, 24, 47, 10, 1988, 20, 22, 1172, 53, 54, 75, 35, 40]
try:
    for i in range(2016,2022):
        try:
            for j in Race_ids:
                try:
                    Race = RaceEdition(race_id=j, year=i)
                    Race = Race.results().results_table
                    Race["Year"] = i
                    Race["Race_ID"] = j
                    races = races.append(Race)
                except: pass
        except: pass
except: pass
```
