```python
races = pd.DataFrame()
Race_ids = [17, 13, 23, 1, 2, 3, 12, 15, 16, 14, 6, 19, 21, 9800, 4, 7, 5, 8, 9, 11, 59, 58, 24, 47, 10, 1988, 20, 22, 1172, 53, 54, 75, 35, 40]
try:
    for i in range(2016,2022):
        try:
            for j in Race_ids:
                try:
                    if(j == 17 or j == 13 or j == 23 or j == 1 or j == 2 or j == 3 or j == 12 or j == 15 or j == 16 or j == 14 or j == 6 or j == 19 or j == 21 or j == 9800):
                        for k in range(1,30):
                            try:
                                Race = RaceEdition(race_id=j, year=i)
                                Race = Race.results(stage_num = k).results_table
                                Race["Year"] = i
                                Race["Stage"] = k
                                Race["Race_ID"] = j
                                races = races.append(Race) 
                            except: pass
                    else:
                        Race = RaceEdition(race_id=j, year=i)
                        Race = Race.results().results_table
                        Race["Year"] = i
                        Race["Stage"] = 0
                        Race["Race_ID"] = j
                        races = races.append(Race)        
                except: pass
        except: pass
except: pass
```
